```python
import seaborn

seaborn.get_dataset_names()
```

```python
diamonds = seaborn.load_dataset('diamonds').head(20)
```

```python
display(diamonds)
```

```python
diamonds_price_cut = diamonds[['cut', 'price']].sort_values(by='price', ascending=False)#.groupby('cut').agg('sum')
```

```python
display(diamonds_price_cut)
```

```python
seaborn.relplot(data=diamonds_price_cut, x='cut', y='price', hue='price')
```
