# 1. Diamantes: preços por tipo  de corte.


```python
import seaborn
```

```python
seaborn.get_dataset_names()
```

```python
diamonds = seaborn.load_dataset('diamonds')
```

```python
diamonds_price_cut = diamonds[['cut', 'price']].groupby('cut').agg('sum').sort_values(by='price', ascending=False)
```

```python
display(diamonds_price_cut)
```

```python
seaborn.relplot(data=diamonds_price_cut, x='cut', y='price', hue='price')
```
